```python
# This file is part of msolve.
#
# msolve is free software: you can redistribute it and/or modify
# it under the terms of the GNU General Public License as published by
# the Free Software Foundation, either version 2 of the License, or
# (at your option) any later version.
#
# msolve is distributed in the hope that it will be useful,
# but WITHOUT ANY WARRANTY; without even the implied warranty of
# MERCHANTABILITY or FITNESS FOR A PARTICULAR PURPOSE.  See the
# GNU General Public License for more details.
#
# You should have received a copy of the GNU General Public License
# along with msolve.  If not, see <https://www.gnu.org/licenses/>
#
# Authors:
# Christian Eder
# Jorge Garcia Fontan
# Huu Phuoc Le
# Mohab Safey El Din
# Bruno Salvy

import os
import sys
import platform
import random
import string
import math
from fractions import Fraction


from helpers import (
    numer,
    expand,
    subs,
    degree,
    SR,
    gcd,
    solve,
    isprime,
    ilog2,
    floor, 
    ceil
)



# ---------------------------------------------------------------------------
# Module-level default path (mirrors Maple's module-local DefaultMSolvePath)
# ---------------------------------------------------------------------------
_DefaultMSolvePath = "msolve"


def GetMSolvePath():
    return _DefaultMSolvePath


def SetMSolvePath(path: str):
    global _DefaultMSolvePath
    _DefaultMSolvePath = path
    return _DefaultMSolvePath


# ---------------------------------------------------------------------------
# Internal helpers
# ---------------------------------------------------------------------------

def GetSystem():
    sys_str = platform.system()
    if sys_str == "Darwin":
        return "macOS"
    elif sys_str == "Linux":
        return "Linux"
    else:
        return "Windows"


def _random_string(n=8):
    return ''.join(random.choices(string.ascii_lowercase, k=n))


def _from_coefficient_list(L, var):
    """Reconstruct polynomial from [length, [c0, c1, ...]] pair."""
    # L is [deg, [c0, c1, ..., cdeg]]
    coeffs = L[1]
    return sum(coeffs[i] * var**i for i in range(len(coeffs)))


def ReadPolynomial(L, var):
    """
    L = [deg, [c0, c1, ..., cdeg]]  (1-indexed in Maple: L[1]=deg, L[2]=coefflist)
    Translated: L[0] = deg+1 count, L[1] = coeff list.
    add(L[2][i]*var^(i-1), i=1..L[1]+1)  -> sum over 0-based index
    """
    n = L[0]  # L[1] in Maple
    coeffs = L[1]  # L[2] in Maple
    return sum(coeffs[i] * var**i for i in range(n + 1))


def ToMSolve(F, char, vars, fname):
    with open(fname, 'w') as fd:
        # Write variable names separated by commas
        for i in range(len(vars) - 1):
            fd.write(f"{vars[i]}, ")
        fd.write(f"{vars[-1]} \n")
        fd.write(f"{char}\n")

        if char == 0:
            F2 = [expand(numer(f)) for f in F]
            F2 = [f for f in F2 if f != 0]
            for i in range(len(F2) - 1):
                fd.write(f"{F2[i]},\n")
            fd.write(f"{F2[-1]}\n")
        else:
            F2 = [expand(numer(f)) for f in F]
            F2 = [f for f in F2 if f != 0]
            for i in range(len(F2) - 1):
                fd.write(f"{F2[i]},\n")
            fd.write(f"{F2[-1]}\n")

    # Join continuation lines with sed (macOS vs Linux syntax)
    if GetSystem() == "macOS":
        str_cmd = f"sed -i '' -e ':a' -e 'N' -e '$!ba' -e 's/\\\\\\n//g' {fname}"
    else:  # Linux
        str_cmd = f"sed -i -e ':a' -e 'N' -e '$!ba' -e 's/\\\\\\n//g' {fname}"
    os.system(str_cmd)


def GetOptions(opts):
    """
    opts: dict mapping option names to values.
    Returns: (str_cmd, fname1, fname2, verb, output)
    """
    # verb
    verb = opts.get("verb", 0)
    if isinstance(verb, int):
        verb = min(verb, 2)
        if verb < 2:
            verb = 0
    else:
        verb = 0

    # gb
    gb_val = opts.get("gb", 0)
    gb = gb_val if (isinstance(gb_val, int) and gb_val > 0) else 0

    # elim
    elim_val = opts.get("elim", 0)
    elim = elim_val if (isinstance(elim_val, int) and elim_val > 0) else 0

    # trunc
    trunc_val = opts.get("trunc", -1)
    truncate = trunc_val if (isinstance(trunc_val, int) and trunc_val > 0) else -1

    # output
    output = opts.get("output", 0)
    if not isinstance(output, int):
        output = 0

    # nthreads
    nthreads = opts.get("nthreads", 1)
    if not isinstance(nthreads, int):
        nthreads = 1

    # linalg
    linalg = opts.get("linalg", 2)
    if not isinstance(linalg, int):
        linalg = 2

    # msolve_path
    if "mspath" in opts:
        msolve_path = opts["mspath"]
        if not isinstance(msolve_path, str):
            print("Error in format options")
            msolve_path = _DefaultMSolvePath
    else:
        msolve_path = _DefaultMSolvePath

    # file_dir
    if "file_dir" in opts:
        file_dir = opts["file_dir"]
        if not isinstance(file_dir, str):
            print("Error in format options")
            file_dir = "/tmp/"
    else:
        file_dir = "/tmp/"

    # fname1 (input file)
    if "file_in" in opts:
        val = opts["file_in"]
        if isinstance(val, str):
            fname1 = os.path.join(file_dir, val)
        else:
            print("Error in format options")
            fname1 = os.path.join(file_dir, _random_string(8) + ".ms")
    else:
        fname1 = os.path.join(file_dir, _random_string(8) + ".ms")

    # fname2 (output file)
    if "file_out" in opts:
        val = opts["file_out"]
        if isinstance(val, str):
            fname2 = os.path.join(file_dir, val)
        else:
            print("Error in format options")
            fname2 = os.path.join(file_dir, _random_string(8) + ".ms")
    else:
        fname2 = os.path.join(file_dir, _random_string(8) + ".ms")

    # param
    param = opts.get("param", 0)
    if not isinstance(param, int):
        param = 0

    # Build the msolve command string
    str_cmd = f"{msolve_path} -v {verb}"
    if not (gb == 0 or param != 0):
        str_cmd += f" -g {gb}"
    if elim != 0:
        str_cmd += f" -e {elim}"
    if param != 0:
        str_cmd += f" -P {param}"
    if truncate != -1:
        str_cmd += f" -N {truncate}"
    if nthreads != 1:
        str_cmd += f" -t {nthreads}"
    if linalg != 2:
        str_cmd += f" -l {linalg}"
    str_cmd += f" -f {fname1} -o {fname2}"

    return str_cmd, fname1, fname2, verb, output


def RemoveFiles(fname1, fname2):
    os.system(f"rm {fname2}")
    os.system(f"rm {fname1}")


def CallMSolve(F, fc, vars, opts):
    str_cmd, fname1, fname2, verb, output = GetOptions(opts)

    nvars = len(vars)
    # Rename variables to internal names _xxi
    xx = list(SR.var('_xx', nvars))
    subs_in  = {vars[i]: xx[i] for i in range(nvars)}
    subs_out = {xx[i]: vars[i] for i in range(nvars)}

    F_renamed = [f.subs(subs_in) for f in F]
    ToMSolve(F_renamed, fc, xx, fname1)

    # Set float precision
    prec = 64  # default (Digits <= 10 in Maple)

    if fc == 0:
        str_cmd = str_cmd + f" -p {prec}"

    try:
        if verb == 0:
            os.system(str_cmd)
        else:
            os.system(str_cmd)
        with open(fname2, 'r') as f:
            raw = f.read()[:-2]
        results_renamed = sage_eval(raw)
        # Substitute internal variable names back
        results = _subs_results(results_renamed, subs_out)
    except Exception as e:
        raise e
        # Write a debug file analogous to /tmp/bug-call-msolve.mpl
        with open("/tmp/bug-call-msolve.py", 'w') as fd:
            fd.write(f"F = {F}\nfc = {fc}\nvars = {vars}\nopts = {opts}\n")
        print(str_cmd)
        raise RuntimeError(
            "There has been an issue in msolve computation (see /tmp/bug-call-msolve.py)"
        ) from e

    RemoveFiles(fname1, fname2)
    return results, output


def _subs_results(obj, subs_map):
    """Recursively apply a substitution dict to a nested structure."""
    if isinstance(obj, list):
        return [_subs_results(x, subs_map) for x in obj]
    if hasattr(obj, 'subs'):
        return obj.subs(subs_map)
    return obj


def CheckCharacteristic(fc):
    if fc != 0 and not isprime(fc):
        raise ValueError("Field characteristic should be a prime number")
    if fc > 2**31:
        raise ValueError("Field characteristic > 2^31 not supported")
    return fc


# ---------------------------------------------------------------------------
# Interval arithmetic helpers
# ---------------------------------------------------------------------------

def NonNegativeIntervalEvaluate(pt, pol):
    """
    pt: list of (var, [lo, hi]) pairs where all intervals are non-negative.
    pol: a polynomial expression.
    Returns [lo, hi] interval containing all values pol can take at pt.
    """
    var_names = [p[0] for p in pt]
    low_pt  = {p[0]: p[1][0] for p in pt}
    up_pt   = {p[0]: p[1][1] for p in pt}

    pol_exp = expand(pol)
    # Collect monomials and coefficients
    pol_dict = pol_exp.as_coefficients_dict()

    pos_terms = {m: c for m, c in pol_dict.items() if c > 0}
    neg_terms = {m: c for m, c in pol_dict.items() if c < 0}

    # Lower bound: positive monomials at low_pt, negative monomials at up_pt
    low_val  = sum(c * m.subs(low_pt) for m, c in pos_terms.items()) \
             + sum(c * m.subs(up_pt)  for m, c in neg_terms.items())
    # Upper bound: positive monomials at up_pt, negative monomials at low_pt
    up_val   = sum(c * m.subs(up_pt)  for m, c in pos_terms.items()) \
             + sum(c * m.subs(low_pt) for m, c in neg_terms.items())

    return [low_val, up_val]


def NonZeroIntervalEvaluate(pt, pol):
    """
    pt: list of (var, [lo, hi]) pairs where none of [lo, hi] contains 0 in interior.
    Returns [lo, hi] interval for pol evaluated at pt.
    """
    if degree(pol) <= 0:
        return [pol, pol]

    newpol = pol
    newpt  = []
    for var, interval in pt:
        if interval[1] < 0:
            newpol = newpol.subs(var, -var)
            newpt.append((var, [-interval[1], -interval[0]]))
        else:
            newpt.append((var, interval))

    return NonNegativeIntervalEvaluate(newpt, newpol)


# ---------------------------------------------------------------------------
# Parametrization helpers
# ---------------------------------------------------------------------------

def ExtractParametrization(res, v):
    """
    res: msolve parametrization result block.
    v: the elimination variable.
    Returns: (param, den, cfs, newres)
    """
    param   = ReadPolynomial(res[0], v)
    den     = ReadPolynomial(res[1], v)
    cfs     = [entry[-1] for entry in res[2]]   # map2(op,-1,res[3]) in Maple
    newres  = [entry[0]  for entry in res[2]]   # map2(op,1,res[3])
    newres  = [ReadPolynomial(r, v) for r in newres]
    return param, den, cfs, newres


def SplitParamCoord(param, pol):
    """
    Split a parametrization block at a factor pol of elim.
    Returns a list of (possibly two) parametrization blocks.
    """
    elim, den, cfs, nums = param[0], param[1], param[2], param[3]

    # Zero out coordinates divisible by pol
    new_nums = [0 if (nums[j] % pol == 0) else nums[j] for j in range(len(nums))]
    param1 = [pol, den, cfs, new_nums]
    lparam = [param1]

    newelim = (elim / pol).cancel()
    if degree(newelim) > 0:
        param2 = [newelim, den, cfs, nums]
        lparam.append(param2)

    return lparam


def Eval_linform(lin_form, sols):
    """
    Evaluate lin_form (a linear combination of variables) at each solution in sols.
    sols: list of lists of (var, [lo, hi]) interval pairs.
    Returns a list of [lo, hi] intervals.
    """
    if len(sols) == 0:
        return []

    lv = lin_form.variables() # TODO
    if len(lv) == 1:
        v = list(lv)[0]
        return [dict(sol)[v] for sol in sols]
    else:
        # lin_form involves an extra variable v not in vars
        vars_in_sols = set()
        for sol in sols:
            for var, _ in sol:
                vars_in_sols.add(var)
        extra = lv - vars_in_sols
        if len(extra) != 1:
            print(lin_form)
            raise ValueError("lin_form incorrect")
        v = list(extra)[0]
        nlin = solve(lin_form, v)[0]
        return [NonZeroIntervalEvaluate(pt, nlin) for pt in sols]


def MakeBinaryInterval(interv):
    lo, hi = interv[0], interv[1]
    p1 = ilog2(lo.denom())
    p2 = ilog2(hi.denom())
    return [floor(interv[1]*2**p1)/2**p1, ceil(interv[2]*2**p2)/2**p2]


def BuildSolution(param, interv, vars):
    """
    Build a solution point (list of (var, [lo, hi]) pairs) from a parametrization
    block and an isolating interval for the elimination polynomial root.
    Also returns boo: True if some coordinate sign is still undetermined.
    """
    v = list(param[0].variables())[0] # TODO
    den  = param[1]
    cfs  = param[2]
    nums = param[3]

    den_interv  = NonZeroIntervalEvaluate([(v, interv)], den)
    nums_interv = [NonZeroIntervalEvaluate([(v, interv)], n) for n in nums]

    coords = []
    for i in range(len(nums_interv)):
        ni = nums_interv[i]
        di = den_interv
        ci = cfs[i]
        lo = ni[0] / (di[1] * ci)
        hi = ni[1] / (di[0] * ci)
        coords.append([lo, hi])
    coords.append(interv)  # last coordinate is the root itself

    # Check which coords contain 0 in their interior (sign undetermined)
    lb = [
        (c[0] <= 0 and c[1] >= 0) and not (c[0] == 0 and c[1] == 0)
        for c in coords
    ]
    boo = any(lb)

    # Build solution as list of (var, [lo, hi]) — flip sign of interval if needed
    if v in vars:
        sol = []
        for i in range(len(vars) - 1):
            c = coords[i]
            if c[1] < 0:
                sol.append((vars[i], [-c[0], -c[1]]))
            else:
                sol.append((vars[i], [-c[1], -c[0]]))
        sol.append((v, interv))
    else:
        sol = []
        for i in range(len(vars)):
            c = coords[i]
            if c[1] < 0:
                sol.append((vars[i], [-c[0], -c[1]]))
            else:
                sol.append((vars[i], [-c[1], -c[0]]))

    return sol, boo


def RefineSolutions(param, lin_form, psols, vars):
    """
    Refine solutions until all coordinate signs are determined.
    """
    elim = param[0]
    v    = list(elim.variables())[0] # TODO
    newsols = []

    for sol in psols:
        interv = MakeBinaryInterval(Eval_linform(lin_form, [sol])[0])
        # Refine root iteratively (placeholder: real impl needs RootFinding)
        # Here we stub with the existing interval; a real port would call
        # a root-refinement routine such as those in mpmath.
        newsol, boo = BuildSolution(param, interv, vars)
        while boo and interv[0] != interv[1]:
            # Refine interval — stub (would use RootFinding.RefineRoot)
            mid = (interv[0] + interv[1]) / 2
            interv = [interv[0], mid] if subs({v:mid}, elim) * subs({v:interv[0]}, elim) <= 0 else [mid, interv[1]]
            newsol, boo = BuildSolution(param, interv, vars)
        newsols.append(newsol)

    return newsols


def SplitAndRefinePerCoordinates(param_in, sols, lb, vars):
    """
    Split the parametrization at coordinate factors and refine solutions
    until all coordinate signs are determined.
    """
    lin_form = param_in[0]
    elim     = param_in[1]
    den      = param_in[2]
    cfs      = param_in[3]
    nums     = param_in[4]

    # Collect indices where at least one lb entry is True
    idx = set()
    for i in range(len(lb)):
        for j in range(len(lb[i])):
            if lb[i][j]:
                idx.add(j)
    idx = sorted(idx)

    lparam = [[elim, den, cfs, nums]]

    for i in idx:
        newlparam = []
        for block in lparam:
            if i == len(vars) - 1 and set(lin_form.variables()) == {vars[-1]}:
                g = gcd(block[0], lin_form)
            else:
                g = gcd(block[0], block[-1][i])
            if degree(g) > 0:
                newlparam.extend(SplitParamCoord(block, g))
            else:
                newlparam.append(block)
        lparam = newlparam

    v    = list(elim.variables())[0]
    lsols = []
    osols = list(sols)

    for param in lparam:
        interv_list = Eval_linform(lin_form, osols)
        vals = [
            [subs({v: iv[0]}, param[0]), subs({v: iv[1]}, param[0])]
            for iv in interv_list
        ]
        psols = []
        for j in range(len(vals)):
            vv = vals[j]
            if (vv[0] == 0 and vv[1] == 0) or (
                (vv[0] < 0) != (vv[1] < 0)
            ):
                psols.append(osols[j])
        osols = [s for s in osols if s not in psols]

        # Update coords known to be zero
        ldeg = [i for i in range(len(param[-1])) if degree(param[-1][i]) < 0]
        for j in ldeg:
            for k in range(len(psols)):
                psols[k][j] = (psols[k][j][0], [0, 0])

        intvl  = [[rhs for _, rhs in pt] for pt in psols]
        newlb  = [
            [(l[0] <= 0 and l[1] >= 0) and not (l[0] == 0 and l[1] == 0) for l in pt]
            for pt in intvl
        ]

        # Remove solutions with no zero-containing coordinate
        psols2 = [psols[j] for j in range(len(psols)) if not any(newlb[j])]
        psols  = [s for s in psols if s not in psols2]

        if any(any(row) for row in newlb):
            psols = RefineSolutions(param, lin_form, psols, vars)

        lsols.extend(psols)
        lsols.extend(psols2)

    return lsols


def SelectVanishingSols(pol, lin_form, sols):
    """
    Return indices of solutions in sols where pol can vanish.
    """
    idx = []
    if degree(pol) == 0:
        return []
    if pol == 0:
        raise ValueError("Input polynomial should not be 0")
    if len(pol.variables()) > 1:
        raise ValueError("Input polynomial should be univariate")

    _v = list(pol.variables())[0]
    interv_list = Eval_linform(lin_form, sols)

    for i, sol_interv in enumerate(interv_list):
        v1 = subs({_v: sol_interv[0]}, pol)
        v2 = subs({_v: sol_interv[1]}, pol)
        s1 = int(v1 > 0) - int(v1 < 0)
        s2 = int(v2 > 0) - int(v2 < 0)
        if s1 * s2 == -1 or v1 == 0 or v2 == 0:
            idx.append(i)

    return idx


# ---------------------------------------------------------------------------
# Parametrization (internal)
# ---------------------------------------------------------------------------

def Parametrization(F, vars, opts):
    """
    Compute rational parametrization of the solutions of F.
    Returns (newvars, lparam, lsols) or (newvars, [0/1]) depending on opts["param"].
    """
    str_val = opts.get("param")
    verb    = opts.get("verb", 0)
    if not isinstance(verb, int):
        verb = 0
    if not isinstance(str_val, Integer):
        raise ValueError(
            "Bad call to Parametrization (opts does not specify the param option)"
        )

    results, output = CallMSolve(F, 0, vars, opts)
    dim = results[0]

    if dim == -1:
        if verb >= 2:
            print("Algebraic set is empty")
        if str_val == 2:
            return vars, [1]
        else:
            return vars, [1], [-1, []]

    if dim > 0:
        if verb >= 2:
            print("Algebraic set has positive dimension")
        return vars, [0], [1, []]

    res = results[1]
    for _k in res:
        print(_k)
        print()
    nvars    = res[1]
    newvars  = res[3]           # 0-based: res[4] in Maple (1-indexed) = res[3]
    v        = newvars[-1]

    _T = SR.var('_T')
    if v not in vars:
        v = _T
        newvars[-1] = _T

    lin_form = sum(res[4][i] * newvars[i] for i in range(len(newvars)))
    res_inner = res[5]
    if res_inner[0] != 1:
        raise ValueError("several output parametrizations")
    res_inner = res_inner[1]

    elim, den, cfs, nums = ExtractParametrization(res_inner, v)

    if str_val == 2:
        return newvars, [lin_form, elim, den, cfs, nums]
    else:
        return newvars, [lin_form, elim, den, cfs, nums], results[2]


def SplitAndRefinePerConstraints(param_in, sols, vals, F, vars, cstr, idx, opts):
    """
    Refine solutions and their constraint values by splitting parametrizations
    at the constraints in cstr.
    """
    lin_form = param_in[0]
    elim     = param_in[1]
    den      = param_in[2]
    cfs      = param_in[3]
    nums     = param_in[4]

    newsols = []
    newvals = []
    tsols   = []
    tvals   = []

    # Separate solutions with ambiguous constraint signs from clear ones
    for i in range(len(sols)):
        ambiguous = any(
            (l[0] <= 0 and l[1] >= 0) and not (l[0] == 0 and l[1] == 0)
            for l in vals[i]
        )
        if ambiguous:
            tvals.append(list(vals[i]))
            tsols.append(sols[i])
        else:
            newvals.append(vals[i])
            newsols.append(sols[i])

    # Compute parametrizations for F ∪ {cstr[i]} for each constraint
    lparam = []
    for i, c in enumerate(cstr):
        newvars_c, param_c = Parametrization([*F, c], vars, {**opts, "param": 2})
        if param_c[0] != 1:  # intersection not empty
            if param_c[0] != lin_form:
                # Linear form changed: need elimination
                lv_lin = lin_form.variables()
                vars_set = set(vars)
                if not lv_lin.issubset(vars_set):
                    pvars = list(vars)
                    vvar  = list(lv_lin - vars_set)[0]
                    pvars.append(vvar)
                    new_lin_form = lin_form
                else:
                    if len(lv_lin) != 1:
                        raise RuntimeError("Bug detected")
                    vvar = list(lv_lin)[0]
                    pvars = [v2 for v2 in vars if v2 != vvar] + [vvar]
                    new_lin_form = None

                gb = MSolveGroebner(
                    [*F, c, new_lin_form] if new_lin_form else [*F, c],
                    0,
                    pvars,
                    {**opts, "elim": len(pvars) - 1}
                )
                param_c[1] = gb[0]

            g = gcd(elim, param_c[1])
            idsols = SelectVanishingSols(g, lin_form, tsols)
            for j in idsols:
                tvals[j][idx[i]] = [0, 0]
            lparam.append(param_c)

    # Re-partition
    newtvals  = []
    newtsols  = []
    for i in range(len(tsols)):
        still_ambiguous = any(
            (l[0] <= 0 and l[1] >= 0) and not (l[0] == 0 and l[1] == 0)
            for l in tvals[i]
        )
        if not still_ambiguous:
            newsols.append(tsols[i])
            newvals.append(tvals[i])
        else:
            newtvals.append(tvals[i])
            newtsols.append(tsols[i])

    if len(newtvals) == 0:
        return newsols, newvals

    # Refine remaining ambiguous solutions
    for i in range(len(newtsols)):
        boo     = False
        newsol  = [newtsols[i]]
        while not boo:
            interv   = MakeBinaryInterval(Eval_linform(lin_form, newsol)[0])
            interv   = interv  # stub: RootFinding.RefineRoot would go here
            newsol_s, boo = BuildSolution([elim, den, cfs, nums], interv, vars)
            newsol   = [newsol_s]
            for j in range(len(newtvals[i])):
                boo = True
                l = newtvals[i][j]
                if (l[0] <= 0 and l[1] >= 0) and not (l[0] == 0 and l[1] == 0):
                    if j not in idx:
                        raise RuntimeError("Bug in refinement")
                    r   = idx.index(j)
                    val = NonZeroIntervalEvaluate(newsol[0], cstr[r])
                    if val[0] <= 0 and val[1] >= 0:
                        boo = False
                        break
                    else:
                        newtvals[i][j] = val
            if boo:
                newsols.append(newtsols[i])
                newvals.append(newtvals[i])

    return newsols, newvals


# ---------------------------------------------------------------------------
# Public API
# ---------------------------------------------------------------------------

def MSolveGroebner(F, fc, vars, opts=None):
    """
    Compute a Groebner basis of the ideal generated by F.

    Input:
      F    : list of polynomials with rational coefficients
      fc   : field characteristic (nonneg int; 0 = rationals)
      vars : list of variables
      opts : dict of options (see comments below)

    Options:
      "mspath"   : path to msolve binary
      "verb"     : verbosity (positive int)
      "file_dir" : directory for intermediate files
      "file_in"  : input filename
      "file_out" : output filename
      "leadmons" : 1 to return only leading monomials
      "elim"     : number of variables to eliminate

    Output:
      []     -> error during computation
      else   -> Groebner basis (grevlex, vars[0] > ... > vars[n-1])
    """
    if opts is None:
        opts = {}
    field_char = CheckCharacteristic(fc)

    merged_opts = {**opts, "gb": 2}
    str_cmd, fname1, fname2, verb, output = GetOptions(merged_opts)

    nvars   = len(vars)
    xx      = list(SR.var('_xx', nvars))
    subs_in  = {vars[i]: xx[i] for i in range(nvars)}
    subs_out = {xx[i]: vars[i] for i in range(nvars)}

    F_renamed = [f.subs(subs_in) for f in F]
    ToMSolve(F_renamed, field_char, xx, fname1)

    try:
        if verb == 0:
            os.system(str_cmd)
        else:
            os.system(str_cmd)
        with open(fname2, 'r') as f:
            raw = f.read()
        results_renamed = eval(raw)
        results = _subs_results(results_renamed, subs_out)
        RemoveFiles(fname1, fname2)
        return results
    except Exception as e:
        raise RuntimeError("There has been an issue in msolve computation") from e


def MSolveGroebnerLM(F, fc, vars, opts=None):
    """
    Like MSolveGroebner but returns only leading monomials (gb=1).
    """
    if opts is None:
        opts = {}
    field_char = CheckCharacteristic(fc)

    merged_opts = {**opts, "gb": 1}
    str_cmd, fname1, fname2, verb, output = GetOptions(merged_opts)

    nvars   = len(vars)
    xx      = SR.var('_xx', nvars)
    subs_in  = {vars[i]: xx[i] for i in range(nvars)}
    subs_out = {xx[i]: vars[i] for i in range(nvars)}

    F_renamed = [f.subs(subs_in) for f in F]
    ToMSolve(F_renamed, field_char, xx, fname1)

    try:
        if verb == 0:
            os.system(str_cmd)
        else:
            os.system(str_cmd)
        with open(fname2, 'r') as f:
            raw = f.read()
        results_renamed = eval(raw)
        results = _subs_results(results_renamed, subs_out)
        RemoveFiles(fname1, fname2)
        return results
    except Exception as e:
        raise RuntimeError("There has been an issue in msolve computation") from e


def MSolveRealRoots(F, vs, G=None, opts=None):
    """
    Compute real roots of the polynomial system F.

    Input:
      F    : list of polynomials with rational coefficients
      vars : list of variables
      G    : list of extra constraint polynomials (can be empty list)
      opts : dict of options

    Options:
      "mspath"   : path to msolve binary
      "verb"     : verbosity
      "file_dir" : directory for intermediate files
      "file_in"  : input filename
      "file_out" : output filename
      "output"   : 1 to return midpoints of isolating intervals

    Output:
      []          -> error
      [1]         -> infinitely many complex solutions
      [-1, []]    -> no real solution
      [0, [sols]] -> finitely many complex solutions; sols is a list of
                     {var: [lo, hi]} dicts (isolating intervals)
      [0, [sols], [vals]]  -> same, plus interval values of G at each solution
      if "output"=1: intervals replaced by midpoints
    """
    if G is None:
        G = []
    if opts is None:
        opts = {}

    verb = opts.get("verb", 0)
    if not isinstance(verb, int):
        verb = 0

    # If any polynomial is a nonzero constant the system has no solution
    if any(degree(f) == 0 for f in F):
        return [-1, []]

    output = opts.get("output", 0)
    if len(G) > 0 and output == 1:
        raise ValueError(
            "mid points cannot be returned with extra constraints: change your options"
        )

    merged_opts = {**opts, "param": 1}
    newvars, lparam, lsols = Parametrization(F, vs, merged_opts)

    if lsols[0] == -1:
        return lsols
    if lparam == [0]:
        return [1]

    nl   = lsols[0]
    sols = []
    for i in range(1, nl + 1):
        point = lsols[i]
        sol   = {newvars[j]: point[j] for j in range(len(vs))}
        sols.append(sol)

    if output == 1:
        sols = [{k: (v[0] + v[1]) / 2 for k, v in sol.items()} for sol in sols]
        return [0, sols]

    # Convert sols to list-of-interval-pair format for internal processing
    intervals = [[rhs for _, rhs in sol.items()] for sol in sols]
    lb = [
        [(l[0] <= 0 and l[1] >= 0) and not (l[0] == 0 and l[1] == 0) for l in pt]
        for pt in intervals
    ]

    if any(any(row) for row in lb):
        sols = SplitAndRefinePerCoordinates(lparam, sols, lb, newvars[:len(vs)])

    if len(G) == 0:
        return [0, sols]

    vals = [
        [NonZeroIntervalEvaluate(list(sol.items()), G[j]) for j in range(len(G))]
        for sol in sols
    ]
    idx = sorted({
        j
        for sol_vals in vals
        for j, l in enumerate(sol_vals)
        if (l[0] <= 0 and l[1] >= 0) and not (l[0] == 0 and l[1] == 0)
    })

    if len(idx) == 0:
        return [0, sols, vals]

    cstr = [G[i] for i in idx]
    sols, vals = SplitAndRefinePerConstraints(
        lparam, sols, vals, F, newvars[:len(vs)], cstr, idx, opts
    )
    return [0, sols, vals]


def MSolveParam(F, fc, vars, opts=None):
    """
    Compute a rational parametrization of the solutions of F.

    Returns [elim_poly, [var_i = expr_i, ...]] where each coordinate
    is expressed as a rational function of the root of elim_poly.
    """
    if opts is None:
        opts = {}

    res = MSolveGroebner(F, fc, vars, {**opts, "param": 2})

    if len(res) == 0:
        raise RuntimeError("There has been an issue in msolve computation")

    dim = res[0]
    if dim == -1:
        raise ValueError("empty set")
    if dim > 0:
        raise ValueError("not 0-dimensional")

    res     = res[1]
    nvars   = res[1]
    newvars = res[3]
    v       = newvars[-1]

    _T = SR.var('_T')
    if v not in vars:
        v = _T

    res_inner = res[5]
    if res_inner[0] != 1:
        raise ValueError("several output parametrizations")
    res_inner = res_inner[1]

    param, den, cfs, nums = ExtractParametrization(res_inner, v)

    return [
        param,
        [newvars[i] == (-nums[i] / (cfs[i] * den)).cancel()
         for i in range(nvars - 1)]
    ]
```

```python
x1, x2, x3, x4, x5, x6 = SR.var('x1 x2 x3 x4 x5 x6')

F = [
    x1 + 2*x2 + 2*x3 + 2*x4 + 2*x5 + 2*x6 - 1,
    x1**2 + 2*x2**2 + 2*x3**2 + 2*x4**2 + 2*x5**2 + 2*x6**2 - x1,
    2*x1*x2 + 2*x2*x3 + 2*x3*x4 + 2*x4*x5 + 2*x5*x6 - x2,
    x2**2 + 2*x1*x3 + 2*x2*x4 + 2*x3*x5 + 2*x4*x6 - x3,
    2*x2*x3 + 2*x1*x4 + 2*x2*x5 + 2*x3*x6 - x4,
    x3**2 + 2*x2*x4 + 2*x1*x5 + 2*x2*x6 - x5,
]
vars_list = [x1, x2, x3, x4, x5, x6]

MSolveRealRoots(F, [x1,x2,x3,x4,x5,x6], [])
```
